```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
address = str(input("Please enter the url: "))
r = requests.get(address)

# create a BeautifulSoup Object
soup = BeautifulSoup(r.text, 'html.parser')
```

```python
def initial_check():
    link_good()
    blocked()
    content_type()
```

```python
def content_type():
    print("Content Type:",end='\n')
    print(r.headers.get("content-type","unknown"),end='\n')
    print('\n')
```

```python
def blocked():
    if "blocked" in r.text:
        print("blocked status:", end='\n')
        print(" we've been blocked", end='\n')
        print('\n')
    else:
        print("blocked status:", end='\n')
        print(" Not blocked", end='\n')
        print('\n')
```

```python

def link_good():
    if r.status_code == 200:
        print("link status:", end='\n')
        print(" The link is goood", end='\n')
        print('\n')
    elif r.status_code == 400 or r.status_code == 500:
        print("link status:", end='\n')
        print(" The link is bad", end='\n')
        print('\n')
```

```python
initial_check()
```

```python
table = soup.find("table",attrs ={'id':'player_statsGrid'})
```

```python
tbody = table.find('tbody')
```

```python
names = tbody.findAll('p')
```

```python
rows = tbody.findAll('tr')
```

```python
data = []
for row in rows:
    cols = row.findAll('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele])
```

```python
df = pd.DataFrame(data, columns=['All Seasons',
                            'Rank',
                            'Player Name',
                            'Team Name', 
                            'GP',
                            'MPG',
                            'PPG',
                            'OBPG',
                            'DBPG',
                            'TotRebs',
                            'Ast',
                            'Stl',
                            'Blk',
                            'TO',
                            'PF',
                            'FGM',
                            'FGA',
                            'FG%',
                            '3PTM',
                            '3PTA',
                            '3PT%',
                            'FTM',
                            'FTA',
                            'FT%'
                           ])
```

```python
df.drop(labels=['All Seasons','Rank'], axis = 1, inplace = True)
```
